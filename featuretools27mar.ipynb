```python
import featuretools as ft
import pandas as pd
```


### Load cleaned data from Wynette's EDA

```python
df = pd.read_csv('clean_dataset.csv')
df
```

### Add Tracker column to keep track of index

```python
df.insert(0, 'Tracker', range(4424))
df
```

### CONVERT 'TARGET' COLUMN from 'CATEGORY' TO 'INTEGER'; Dropout:1, Graduate:3, Enrolled:2

```python

df['Target'] = df['Target'].astype('category').cat.codes + 1
df
```

### Creating target_df which acts as child_dataframe; plus we want to derive features for gradstatus

```python
target_df = pd.DataFrame(df['Target'])
target_df.insert(0, 'Tracker', range(4424))
target_df
```

### Create a dictionary with the DataFrames in dataset; DataFrames are passed in with their index column

```python
dataframes = {
    "wholedata": (df, "Tracker"),
    "gradstatus": (target_df, "Tracker"),
}
```

### Create relationships between dataframes: (parent_dataframe,parent_column,child_dataframe,child_column)

```python
relationships = [
    ("wholedata", "Tracker", "gradstatus", "Target"),
]
```

### Run Deep Feature Synthesis (DFS)

```python
feature_matrix_gradstatus, feature_defs = ft.dfs(
    dataframes=dataframes,
    relationships=relationships,
    target_dataframe_name="gradstatus",)
feature_matrix_gradstatus
```

### Change target DataFrame: DFS can create a feature matrix for any DataFrame in our set

```python
feature_matrix_wholedata, features_defs = ft.dfs(
    dataframes=dataframes,
    relationships=relationships,
    target_dataframe_name="wholedata",)
feature_matrix_wholedata
```
